```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
```

```python
def plot_losses(losses):
    # do a line plot for train_node_loss and val_node_loss over the epochs
    # do a line plot for train_net_loss and val_net_loss over the epochs
    # create a subplot with 2 rows and 1 column
    fig, ax = plt.subplots(2, 1, figsize=(10, 10))
    # plot the node loss
    losses[['train_node_loss', 'val_node_loss']].plot(ax=ax[0])
    ax[0].set_title('Node Loss')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[0].legend(['Train', 'Validation'])
    # plot the net loss
    losses[['train_net_loss', 'val_net_loss']].plot(ax=ax[1])
    ax[1].set_title('Net Loss')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Loss')
    ax[1].legend(['Train', 'Validation'])
    plt.savefig('loss_curves.png')
    plt.show()
```

```python
losses_baseline = pd.read_csv('../profiling_results/baseline/losses.csv')
losses_baseline = losses_baseline.set_index('epoch')
plot_losses(losses_baseline)
```

```python
losses_baseline_earlystop = pd.read_csv('../profiling_results/baseline-earlystop/performance.csv')
losses_baseline_earlystop = losses_baseline_earlystop.set_index('epoch')
plot_losses(losses_baseline_earlystop)
```

```python
losses_baseline_earlystop_reducedim = pd.read_csv('../profiling_results/baseline-earlystop-reducedim/performance.csv')
losses_baseline_earlystop_reducedim = losses_baseline_earlystop_reducedim.set_index('epoch')
plot_losses(losses_baseline_earlystop_reducedim)
```

```python
losses_baseline_earlystop_reduceall = pd.read_csv('../profiling_results/baseline-earlystop-reduceall/performance.csv')
losses_baseline_earlystop_reduceall = losses_baseline_earlystop_reduceall.set_index('epoch')
plot_losses(losses_baseline_earlystop_reduceall)
```

```python
# calculate the min, max, and mean of gpu_peak_memory for each model
def gpu_peak_memory_stats(performance):
    df = pd.read_csv(performance)
    df = df['gpu_peak_memory']
    print(f"Min GPU Peak Memory: {df.min()}")
    print(f"Max GPU Peak Memory: {df.max()}")
    print(f"Mean GPU Peak Memory: {df.mean()}")

print("Baseline with Early Stopping")
gpu_peak_memory_stats('baseline-earlystop/performance.csv')
print("Early Stopping with Reduced Dimensionality")
gpu_peak_memory_stats('baseline-earlystop-reducedim/performance.csv')
print("Early Stopping with Reduced Layers")
gpu_peak_memory_stats('baseline-earlystop-reducelayers/performance.csv')
print("Early Stopping with Reduced Dimensionality and Reduced Layers")
gpu_peak_memory_stats('baseline-earlystop-reduceall/performance.csv')
print("Custom Learning Rate and Reduced Dimensionality and Reduced Layers")
gpu_peak_memory_stats('custom-reduceall/performance.csv')
print("Custom Learning Rate and Reduced Dimensionality and Reduced Layers and Adjusted Early Stopping")
gpu_peak_memory_stats('custom2_reduceall/performance.csv')
```
