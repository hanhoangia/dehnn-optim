```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_losses(losses, title):
    # do a line plot for train_node_loss and val_node_loss over the epochs
    # do a line plot for train_net_loss and val_net_loss over the epochs
    # create a subplot with 2 rows and 1 column
    fig, ax = plt.subplots(2, 1, figsize=(10, 10))
    # set the title of the plot
    fig.suptitle(title)
    # plot the node loss
    losses[['train_node_loss', 'val_node_loss']].plot(ax=ax[0])
    ax[0].set_title('Node Loss')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[0].legend(['Train', 'Validation'])
    # plot the net loss
    losses[['train_net_loss', 'val_net_loss']].plot(ax=ax[1])
    ax[1].set_title('Net Loss')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Loss')
    ax[1].legend(['Train', 'Validation'])
```

```python
losses_baseline = pd.read_csv('../profiling_results/baseline/performance.csv')
losses_baseline = losses_baseline.set_index('epoch')
losses_baseline_earlystop = pd.read_csv('../profiling_results/baseline-earlystop/other_metrics.csv')
# get the stop epoch
stop_epoch = losses_baseline_earlystop['stop_epoch'].values[0]
# plot the losses
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
# set the title of the plot
fig.suptitle('Baseline with Early Stopping')
# plot the node loss
losses_baseline[['train_node_loss', 'val_node_loss']].plot(ax=ax[0])
ax[0].set_title('Node Loss')
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Loss')
ax[0].legend(['Train', 'Validation'])

# add a vertical line at the stop epoch
ax[0].axvline(x=stop_epoch, color='r', linestyle='--', label=f'Stop Epoch = {stop_epoch}')
ax[0].text(stop_epoch + 1, ax[0].get_ylim()[1] * 0.9, f'Stop Epoch = {stop_epoch}', color='r', fontsize=18, va='center')
# plot the net loss
losses_baseline[['train_net_loss', 'val_net_loss']].plot(ax=ax[1])
ax[1].set_title('Net Loss')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Loss')
ax[1].legend(['Train', 'Validation'])
# plot the stop epoch with annotation 'Stop Epoch = {stop_epoch}'

ax[1].axvline(x=stop_epoch, color='r', linestyle='--', label=f'Stop Epoch = {stop_epoch}')
ax[1].text(stop_epoch + 1, ax[1].get_ylim()[1] * 0.9, f'Stop Epoch = {stop_epoch}', color='r', fontsize=18, va='center')
plt.savefig('../profiling_results/baseline/losses_curves_with_stop_epoch.png')
```

```python
losses_baseline = pd.read_csv('../profiling_results/baseline/performance.csv')
losses_baseline = losses_baseline.set_index('epoch')
plot_losses(losses_baseline, 'Baseline')
plt.savefig('../profiling_results/baseline/loss_curves.png')
plt.show()
```

```python
losses_baseline_earlystop = pd.read_csv('../profiling_results/baseline-earlystop/performance.csv')
losses_baseline_earlystop = losses_baseline_earlystop.set_index('epoch')
plot_losses(losses_baseline_earlystop, 'Baseline Earlystop Losses')
plt.savefig('../profiling_results/baseline-earlystop/loss_curves.png')
plt.show()
```

```python
losses_baseline_earlystop_aa = pd.read_csv('../profiling_results/earlystop-architectureadjustment/performance.csv')
losses_baseline_earlystop_aa = losses_baseline_earlystop_aa.set_index('epoch')
plot_losses(losses_baseline_earlystop_aa, 'Earlystop + Architecture Adjustment Model')
plt.savefig('../profiling_results/earlystop-architectureadjustment/loss_curves.png')
plt.show()
```
