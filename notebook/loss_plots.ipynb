```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os
```

```python
def plot_losses(losses, title):
    # do a line plot for train_node_loss and val_node_loss over the epochs
    # do a line plot for train_net_loss and val_net_loss over the epochs
    # create a subplot with 2 rows and 1 column
    fig, ax = plt.subplots(2, 1, figsize=(10, 10))
    # set the title of the plot
    fig.suptitle(title)
    # plot the node loss
    losses[['train_node_loss', 'val_node_loss']].plot(ax=ax[0])
    ax[0].set_title('Node Loss')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[0].legend(['Train', 'Validation'])
    # plot the net loss
    losses[['train_net_loss', 'val_net_loss']].plot(ax=ax[1])
    ax[1].set_title('Net Loss')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Loss')
    ax[1].legend(['Train', 'Validation'])
```

```python
losses_baseline = pd.read_csv('../profiling_results/baseline/performance.csv')
losses_baseline = losses_baseline.set_index('epoch')
plot_losses(losses_baseline, 'Baseline Losses')
plt.savefig('../profiling_results/baseline/loss_curves.png')
plt.show()
```

```python
losses_baseline_earlystop = pd.read_csv('../profiling_results/baseline-earlystop/performance.csv')
losses_baseline_earlystop = losses_baseline_earlystop.set_index('epoch')
plot_losses(losses_baseline_earlystop, 'Baseline Earlystop Losses')
plt.savefig('../profiling_results/baseline-earlystop/loss_curves.png')
plt.show()
```

```python
losses_baseline_earlystop_aa = pd.read_csv('../profiling_results/earlystop-architectureadjustment/performance.csv')
losses_baseline_earlystop_aa = losses_baseline_earlystop_aa.set_index('epoch')
plot_losses(losses_baseline_earlystop_aa, 'Earlystop + Architecture Adjustment Model Losses')
plt.savefig('../profiling_results/earlystop-architectureadjustment/loss_curves.png')
plt.show()
```

```python
losses_optimized = pd.read_csv('../profiling_results/optimized/performance.csv')
losses_opitmized = losses_optimized.set_index('epoch')
plot_losses(losses_optimized, 'Optimized Model Losses')
plt.savefig('../profiling_results/optimized/loss_curves.png')
plt.show()
```
