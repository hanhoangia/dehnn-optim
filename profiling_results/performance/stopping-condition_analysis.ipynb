```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
losses = pd.read_csv('2025-02-20_01-50-36/losses.csv')
# make epoch the index
losses = losses.set_index('epoch')
losses
```

```python
# plot train_node_loss and val_node_loss on the same graph using line plot
losses[['train_node_loss', 'val_node_loss']].plot()
plt.title('Train and Validation Node Loss for Configuration 2')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.savefig('loss_curve_2.png')
plt.show()
```

```python
losses_rebench = pd.read_csv('losses_rebenchmark.csv')
# make epoch the index
losses_rebench = losses_rebench.set_index('epoch')

# plot train_node_loss and val_node_loss on the same graph using line plot
# plot train_net_loss and val_net_loss on the same graph using line plot
# using subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))
losses_rebench[['train_node_loss', 'val_node_loss']].plot(ax=axs[0])
axs[0].set_title('Train and Validation Node Loss')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')
losses_rebench[['train_net_loss', 'val_net_loss']].plot(ax=axs[1])
axs[1].set_title('Train and Validation Net Loss')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')
plt.tight_layout()
plt.savefig('loss_curve_rebench.png')
plt.show()
```

```python
losses_earlystop = pd.read_csv('losses_earlystop.csv')
# make epoch the index
losses_earlystop = losses_earlystop.set_index('epoch')

# plot train_node_loss and val_node_loss on the same graph using line plot
# plot train_net_loss and val_net_loss on the same graph using line plot
# using subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))
losses_earlystop[['train_node_loss', 'val_node_loss']].plot(ax=axs[0])
axs[0].set_title('Train and Validation Node Loss')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Loss')
losses_earlystop[['train_net_loss', 'val_net_loss']].plot(ax=axs[1])
axs[1].set_title('Train and Validation Net Loss')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')
plt.tight_layout()
plt.savefig('loss_curve_earlystop.png')
plt.show()
```
