```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
losses = pd.read_csv('losses.csv')
# make epoch the index
losses = losses.set_index('epoch')
losses
```

```python
# plot train_node_loss and val_node_loss on the same graph using line plot
losses[['train_node_loss', 'val_node_loss']].plot()
plt.title('Train and Validation Node Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```
